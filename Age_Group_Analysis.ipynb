## **Analysis of Drug Cnsumption by Age Group** 

**My Hypothesis:** 
- Illicit drug use is most common among young adults and decreases as respondents age

**My Question:**
- Does illicit drug consumption reduce as we age?

introduction: Using data records of Drug Consumption Classification we analysed for certain features of the data.  My role was to analyse the age groups of respondents and investigate if drug consumption reduces as we age.

### **Analysis**

Stacked Bar Chart analysis showed - 
- Cannabis respondents to have the most active consumption profiles across all the age groups.  Cannabis appears to be the most commonly used drug.
- Every drug had a consumption profile of being used in the last 10 years across most age groups except the 65+ years age group for all drug types and the 55 to 64 year age group for Ketamine.
- Amphetamine, Cannabis, Cocaine and Meth were overall the most commonly used in the last day, with Cocaine having the least total percentage per age group.
- When focusing on the 65+ years age group we see VSA, Ketamine, Heroin, Crack and Amphetamines falling into the Never Used consumption group.
- However, if we look directly at LSD consumption and Mushroom consumption we see the 65+ years age group show consumption over a decade ago.  This would suggest that in their middle age groupings these drugs where consumed.
- If we look at the stacked bar charts we see a trend of drug use regressing as the respondents age increases.  This is very suggestive of the hypothesis that drug consumption decreases as respondents age.  However, again the limitation of respondent numbers in the 65+ years age group, 18 respondents,  and the 55 to 64 years age group, 91 respondents, compared to the sample size of the other age groups. 592, 442, 339 and 285 respectively, this may skew the data somewhat, however percentages calculated were on the total number of respondents per drug, per age group
- There are potentially 2 drug types as outliers here though,  that being Crack and Amyl Nitrate, as the overall consumption here in the 18 to 24 years age group was less than the 25 to 54 year age groups, however this may be due to generational differences as these drugs were popular in the 70’s to 90’s.

The Scatter Plot analysis of drug consumption across all age groups in the last year only suggests the trend supports the hypothesis that drug consumption decreases with age.  The 4 drug types represented in this chart all have consumption profiles of use in the last day, last week, last month, last year and last decade across most of the age groups with Cocaine being the only drug type at years 45 to 54 years not to have a, used in the last day, profile.   The highest percentage of use was at the 18 to 24 years age group were we see 27 % of 592 respondents used Ecstasy compared with the 25 to 34 years age group where only 13 % 442 of respondents used Ecstasy.  

While this data shows a trend to support the hypothesis that drug use decreases with age we need to consider the sample size per age group.  Perhaps the removal of 65+ years, and the 55 to 64 years age groups could have allowed for a more definitive trend strongly supporting the hypothesis.  

The limitations of the data for an analysis of drug consumption per age group is that this data is only a day in time.  There was not enough numerically available data to present a definitive case for the hypothesis.  If data was collected on the respondents drug consumption, including the type of drug they used, and why they used it, over a period of a year or more we would have a continuum of numerical data types to compare and contrast, over a period of time, as well as trends of use in type of drug;  to show consistent patterns of drug consumption and type of drugs used across age groups, if one should exist.

Another consideration to include considering this data is current, is the use of Cannabis for the treatment of pain.  This could account for the consistent consumption data at the 45 – 54 years age group, and the 55 to 64 years age group.

Conclusion: The trends in this data show that illicit drug use does decrease with age.

## Data Visualisation


### Stacked Bar Chart analysis showing the last time respondents used a certain drug, divided into age groups

#### Stacked Bar Chart generation

```python
#Create stacked bar charts for each illict drug type showing the consumption type for each age group using a for loop

drugs_to_analyse = ['Amphetamines',	'Amyl Nitrite',	'Cannabis',	'Cocaine', 'Crack', 'Ecstasy', 'Heroin','Ketamine',	'LSD',	'Meth',	'Mushrooms','VSA']
age_to_analyse = []
df = pd.DataFrame()


for drug in drugs_to_analyse:
    df = new_data[~new_data["Age"].isin(age_to_analyse)]
    df = df.groupby(["Age", drug]).count()['ID'].unstack()

    percent_df = pd.DataFrame()
    df['Sum'] = df.apply(lambda x: x.sum(), axis=1)
    
    for columns in ['CL0', 'CL1', 'CL2', 'CL3', 'CL4', 'CL5', 'CL6']:
        percent_df[columns] = df.apply(lambda x: 100*x[columns]/x['Sum'], axis=1)


    percent_df.plot(figsize=(12,8),kind='barh', stacked=True)


    plt.xlabel('%')
    plt.legend([CL0, CL1, CL2, CL3, CL4, CL5, CL6], [
    'Never Used', 'Used over a Decade Ago', 'Used in Last Decade', 'Used in Last Year', 'Used in Last Month',
    'Used in Last Week', 'Used in Last Day']);
   
    plt.title(f"% of Total Respondents {drug} Consumption by Age Group")
```
